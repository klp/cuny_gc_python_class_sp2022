```python
# Starting with conditions 
# room_type is Entire home/apt, and within price range
df2 = df[(df['room_type'] == 'Entire home/apt') & (df['price'] >= 30) & (df['price'] <= 500)]
```

```python
# Split into three neighborhood
harlem = df2[df2['neighbourhood'] == 'Harlem']
hkit = df2[df2['neighbourhood'] == "Hell's Kitchen"]
bushwick = df2[df2['neighbourhood'] == 'Bushwick']
```

```python
harlem
```

```python
print(harlem.price.mean())
print(hkit.price.mean())
print(bushwick.price.mean())
```

```python
habukt = pd.concat([harlem, hkit, bushwick])
```

```python
habukt
```

```python
habukt.groupby('neighbourhood').price.mean().plot.bar()
```

```python
plot = habukt.groupby('neighbourhood').price.mean().plot(
    title="Mean price by neighbourhood group",
    color="purple",
    kind='barh',
    xlim=[30, 250],
    xlabel='hoods',
    ylabel=('dollars per day')
).set_xlabel('mean price')
```

```python
fig = plot.get_figure()
fig.savefig('mean_price_per_neighborhood.jpg', dpi=300, bbox_inches = "tight")
```

## Creating column based on condition

```python
# low rent, medium rent, high rent
# We need to use numpy
import numpy as np

# np.select takes conditions then values
conditions = [
    (df.price < 50),
    (df.price > 50) & (df.price < 150),
    (df.price > 150)
]
values = ['low', 'medium', 'high']

df['price_category'] = np.select(conditions, values)
```

```python
df.head()
```

```python
df.groupby('price_category').size().loc[['low', 'medium', 'high']].plot.bar()
```

```python
df.groupby(['neighbourhood_group', 'price_category']).size()
```
